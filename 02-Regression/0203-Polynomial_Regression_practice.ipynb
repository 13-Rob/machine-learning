# Polynomial Linear Regression

---

## Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```


---

## Import the dataset

```python
dataset = pd.read_csv('0200-Data/Position_Salaries.csv')
```

```python
X = dataset.iloc[:, 1:-1].values
```

```python
y = dataset.iloc[:, -1].values
```


---

## Train Linear Regression model

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X, y)
```


---

## Train Polynomial Regression model

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree = 4)
```

```python
X_poly = poly_reg.fit_transform(X, y)
```

```python
lin_reg_2 = LinearRegression()
```

```python
lin_reg_2.fit(X_poly, y)
```


---

## Visualize the results

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg.predict(X))

plt.title('Salary vs Level (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```

```python
plt.scatter(X, y, color='red')
plt.plot(X, lin_reg_2.predict(X_poly))

plt.title('Salary vs Level (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```

```python
X_grid = np.linspace(min(X), max(X))
X_grid = X_grid.reshape((len(X_grid), 1))

plt.scatter(X, y, color='red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)))

plt.title('Salary vs Level (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')

plt.show()
```


---

## Make a single prediction

### Linear regression

```python
lin_reg.predict([[6.5]])
```

### Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
