## Bag-of-Words

The Bag-of-Words uses a vector, and every position in the vector represents a word, or a symbol. The first two positions are reserved for SOS (Start of Sentence) and EOS (End of Sentence), and the last one for special words and words that doesn't fit anywhere else.

To analyze a text you would count every word aparition and place the count on it's corresponding place in the vector.

The rest of the training is similar to other Machine Learning models. You would use a test set, and train the model with that, that way it will be able to give a response (Yes/No) to the email.
